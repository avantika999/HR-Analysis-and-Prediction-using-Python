```python
# showing a correlation map for all numeric variables.
corr_matrix = df.select_dtypes(include=np.number).corr()
plt.figure(figsize=(20,10))
sns.heatmap(corr_matrix,annot=True,cmap="coolwarm")
plt.title('Correlation Map for Numeric Variables')
plt.show()
```

## **Salary Analysis**

```python
# Monthly Income Ditribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='MonthlyIncome', bins=20, kde=True, color='black', edgecolor='purple')
plt.title('Monthly Income Distribution')
plt.xlabel('Monthly Income')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Salary by Jobrole
plt.figure(figsize=(12, 6))
sns.boxplot(x='JobRole', y='MonthlyIncome', data=df)
plt.title('Salary by Job Role')
plt.xticks(rotation=45, ha='right')  # Adjust rotation for better readability
plt.xlabel('Job Role')
plt.ylabel('Monthly Income')
plt.show()
```

## **Employee Satisfaction and Performance**

```python
# Job Satisfaction Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='JobSatisfaction', bins=20, kde=True, color='lightgreen', edgecolor='darkgreen')
plt.title('Job Satisfaction Distribution')
plt.xlabel('Job')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Performance Rating Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='PerformanceRating', bins=20, kde=True, color='Maroon', edgecolor='black')
plt.title('Performance Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Performance')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Employee Tenure
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='YearsAtCompany', bins=20, kde=True, color='grey', edgecolor='black')
plt.title('Employee Tenure Distribution')
plt.xlabel('Employee Tenure')
plt.ylabel('Performance')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Age Distribution by Department
plt.figure(figsize=(10,10))
sns.boxplot(data=df, x='Department',y='Age')
plt.title('Age Distribution by Department')
plt.show()
```

```python
# Attrition by JobRole
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='JobRole',hue='Attrition')
plt.title('Attrition by Job Role')
plt.xticks(rotation=45, ha='right')  # Adjust rotation for better readability
plt.show()
```
